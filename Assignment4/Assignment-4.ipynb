# <span style="color:Green">P1: Load the dataset and perform data preprocessing tasks to save the dataset as csv format in the working directory.</span>

```python
#Codes of P1 here
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

textFile = open("boston.txt", "r")
file_content = textFile.read()
contentList = file_content.split()
textFile.close()

ndarray2D = np.array(contentList)
ndarray2D
finalFileArray = np.reshape(ndarray2D, (-1,14))
#This final array has 14 columns for the 14 variables

dfFile = pd.DataFrame(finalFileArray)
dfFile.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

# SAVE THE DATAFRAME TO A CSV FILE
dfFile.to_csv('boston-dataset.csv', index=False)
```

# <span style="color:Green">P2: Load the csv file from the working directory.</span>

```python
#Codes of P2 here

dataFrame = pd.read_csv('boston-dataset.csv')
```

# <span style="color:Green">P7: Predict the nitrous oxide level (NOX) (apply linear regression for RM vs NOX, PTRATIO vs NOX, LSTAT vs NOX, and MEDV vs NOX).</span>

```python
#Codes of P7 here
from sklearn.linear_model import LinearRegression

# Applying linear regression for RM vs NOX (RM IS Y-VARIABLE, NOX IS X-VARIABLE)
RM = dataFrame['RM'].values.reshape(-1,1)
NOX = dataFrame['NOX'].values.reshape(-1,1)
lr = LinearRegression(fit_intercept=True)
lr.fit(NOX, RM)
Ypred = lr.predict(NOX)

plt.scatter(NOX, RM)
plt.plot(NOX, Ypred, color='red')
plt.title("RM vs NOX")
plt.show()

# Applying linear regression for PTRATIO vs NOX (PTRATIO IS Y-VARIABLE, NOX IS X-VARIABLE)
PTRATIO = dataFrame['PTRATIO'].values.reshape(-1,1)
lr2 = LinearRegression(fit_intercept=True)
lr2.fit(NOX, PTRATIO)
Ypred2 = lr2.predict(NOX)

plt.scatter(NOX, PTRATIO)
plt.plot(NOX, Ypred2, color="blue")
plt.title("PTRATIO vs NOX")
plt.show()

# Applying linear regression for LSTAT vs NOX (LSTAT IS Y-VARIABLE, NOX IS X-VARIABLE)
LSTAT = dataFrame['LSTAT'].values.reshape(-1,1)
lr3 = LinearRegression(fit_intercept=True)
lr3.fit(NOX, LSTAT)
Ypred3 = lr3.predict(NOX)

plt.scatter(NOX, LSTAT)
plt.plot(NOX, Ypred3, color="orange")
plt.title("LSTAT vs NOX")
plt.show()

# Applying linear regression for LSTAT vs NOX (MEDV IS Y-VARIABLE, NOX IS X-VARIABLE)
MEDV = dataFrame['MEDV'].values.reshape(-1,1)
lr4 = LinearRegression(fit_intercept=True)
lr4.fit(NOX, MEDV)
Ypred4 = lr4.predict(NOX)

plt.scatter(NOX, MEDV)
plt.plot(NOX, Ypred4, color="green")
plt.title("MEDV vs NOX")
plt.show()
```

# <span style="color:Green">P8: Predict the price of a home (apply linear regression for RM vs MEDV, PTRATIO vs MEDV, LSTAT vs MEDV, and NOX vs MEDV).</span>

```python
#Codes of P8 here
# Applying linear regression for RM vs MEDV (RM IS Y-VARIABLE, MEDV IS X-VARIABLE)
model1 = LinearRegression(fit_intercept=True)
model1.fit(MEDV, RM)
yfit1 = model1.predict(MEDV)

plt.scatter(MEDV, RM)
plt.plot(MEDV, yfit1, color='red')
plt.title("RM vs MEDV")
plt.show()

# Applying linear regression for PTRATIO vs MEDV (PTRATIO IS Y-VARIABLE, MEDV IS X-VARIABLE)
model2 = LinearRegression(fit_intercept=True)
model2.fit(MEDV, PTRATIO)
yfit2 = model2.predict(MEDV)

plt.scatter(MEDV, PTRATIO)
plt.plot(MEDV, yfit2, color='BLUE')
plt.title("PTRATIO vs MEDV")
plt.show()

# Applying linear regression for LSTAT vs MEDV (LSTAT IS Y-VARIABLE, MEDV IS X-VARIABLE)
model3 = LinearRegression(fit_intercept=True)
model3.fit(MEDV, LSTAT)
yfit3 = model3.predict(MEDV)

plt.scatter(MEDV, LSTAT)
plt.plot(MEDV, yfit3, color='GREEN')
plt.title("LSTAT vs MEDV")
plt.show()

# Applying linear regression for NOX vs MEDV (NOX IS Y-VARIABLE, MEDV IS X-VARIABLE)
model4 = LinearRegression(fit_intercept=True)
model4.fit(MEDV, NOX)
yfit4 = model4.predict(MEDV)

plt.scatter(MEDV, NOX)
plt.plot(MEDV, yfit4, color='BROWN')
plt.title("NOX vs MEDV")
plt.show()
```

# <span style="color:Green">P9: Calculate the coefficient of determination, $R^2$, to quantify the models performance.</span>

```python
#Codes of P9 here
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Printing the RM vs MEDV model
plt.scatter(MEDV, RM)
plt.plot(MEDV, yfit1, color='red')
plt.title("RM vs MEDV")
plt.show()

# I will test the RM vs MEDV model done above
X = MEDV    # MEDV is the x-variable
y = RM      # RM is the y-variable

# First we split the data with 80% training data
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0, train_size=0.8)

# Applying linear regression for RM vs MEDV (RM IS Y-VARIABLE, MEDV IS X-VARIABLE)
model = LinearRegression(fit_intercept=True)
model.fit(Xtrain, ytrain)
r_squared_score = model.score(Xtest, ytest)
print("The R-squared score is", r_squared_score)


```

# <span style="color:Green">P10: How do you evaluate your models performance.</span>

```python
# Discussion of P10 here
# Finding the coefficients
print("Intercept:", model.intercept_[0])
print("coefficient:", model.coef_[0][0])
print("A unit increase in MEDV (x-variable) is associated with a", model.coef_[0][0], "unit increase in RM")

# Manually predicting
# y = B0 + B1x
# y = 5.103049863790971 + 0.053014039651944934 * 45
# y = 7.488681648128493

X_new = pd.DataFrame({'Y_new': [45]})
# Predict for a new observation with x=45
model.predict(X_new)
```
